## Método `.groupby()`:

```python
# documetación del método .groupby()
#help(pd.DataFrame.groupby)
```

### Agrupando por una sola columna:

```python
df.columns
```

```python
# agrupando por una columna
df_agrupado_probs = df.groupby(by=['destino_provincia'], as_index=True)
```

```python
# accediendo a ciertas columnas luego de la agrupación
df_agrupado_probs_loc_personas = df_agrupado_probs[['destino_localidad', 'cantidad_personas']]
```

```python
# viendo los resultados para un grupo en particular con .get_group()
df_agrupado_probs_loc_personas.get_group('Tucumán')
```

```python
# equivalencia de .get_group() utilizando mascaras
```

```python
# aplicando una función de agregación
df_agrupado_probs_loc_personas.agg(func={'cantidad_personas':'mean'})
```

### Agrupando por mas de una columna:

```python
# agrupando por dos columnas

df_prov_loc = df.groupby(by=['destino_provincia', 'destino_localidad'], as_index=True)
```

```python
# accediendo a ciertas columnas luego de la agrupación

df_prov_loc_cp_cc = df_prov_loc[['cantidad_personas', 'cantidad_certificados']]
```

```python
# accediendo a ciertos grupos de las columnas agrupadas luego de la agrupación

df_prov_loc_cp_cc.get_group(('Salta', 'Aguaray'))
```

```python
# Aplicando una función de agregación sobre el dataframe agrupado y con columnas ya seleccionadas
df_prov_loc_cp_cc.agg(func={'cantidad_personas':'sum', 'cantidad_certificados':'mean'})
```

```python
# funcion con argumentos y retorno de info
def cuadrado(col_num):
  return col_num ** 2

df['cantidad_personas'].apply(func=cuadrado)
```

```python
# funcion con argumentos y valores por defecto
# uniendo los valores de dos columnas con apply y una funcion definida
def unir_columnas_str(fila):
  col1 = fila['destino_provincia']
  col2 = fila['destino_localidad']
  union = f"{col1}, {col2}"
  return union

df.apply(func=unir_columnas_str, axis=1)
```

```python
# para obtener valores unicos de un iterable (fila, columna)
df.apply(lambda x: x.unique())
```

```python
# pseudo invocación de una función anonima
cuadrado = lambda x: x ** 2
cuadrado(x=2)
```
