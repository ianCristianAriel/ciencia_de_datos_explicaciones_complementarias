```python
prov_enc.shape
```

```python
# obtener nombre de las columnas de salida
columnas_codificadas = enc_oh.get_feature_names_out()
```

```python
columnas_codificadas.shape
```

```python
df_procesado.tail()
```

```python
# crear dataframe
df_prov_enc = pd.DataFrame(data=prov_enc, columns=columnas_codificadas)
```
